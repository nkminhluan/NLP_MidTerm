# Đọc dữ liệu file train

```python
with open('vi_train.pos',encoding='utf-8') as f:
    f_train = f.readlines()
print(len(f_train))
```

```python
print(f_train)
```

```python
import re
word_train = re.findall(r'(\w+|\s+|\...|[A-Za-z-“”:()?;%>><</~*&^.,!-]+)/', str(f_train))
print(word_train)

```

```python
tag_train = re.findall(r'(\w+|\s+|\...|[A-Za-z-“”:()?;%>><</~*&^.,!-]+)'+r'([/]\w+|[/]\-|[/]\,|[/]\.)',str(f_train))
```

```python
import pandas as pd
df_train = pd.DataFrame(tag_train, columns=['Word', 'Tag']) 
df_train
```

```python
reviewsSVM = df_train.values[0:10000,0]
labelSVM = df_train.values[0:10000,1]
print(reviewsSVM[:5])
print(labelSVM[:5])
```

```python
import nltk
sentencesSVM = [nltk.word_tokenize(sent) for sent in reviewsSVM]
print(sentencesSVM)
```

```python
from keras.preprocessing.text import Tokenizer 
tokenizerSVM = Tokenizer(num_words=None)
tokenizerSVM.fit_on_texts(sentencesSVM)
print(tokenizerSVM.word_index)
```

```python
X_SVM = tokenizerSVM.texts_to_matrix(sentencesSVM,mode='binary')
print(X_SVM)
```

```python
X_count = tokenizerSVM.texts_to_matrix(sentencesSVM, mode = 'count')
print(X_count)
```

```python
from sklearn.model_selection import train_test_split
X_trainSVM, X_testSVM, y_trainSVM, y_testSVM = train_test_split(X_SVM, labelSVM, test_size=0.3, random_state=0)
```

```python
print(X_trainSVM)
```

```python
from sklearn.metrics import classification_report
from sklearn.svm import SVC
# Train mô hình SVM
svm=SVC(C=2, kernel='rbf',gamma='scale')
svm.fit(X_trainSVM, y_trainSVM)
y_svm=svm.predict(X_testSVM)
print(y_svm[0:50])
print(classification_report(y_testSVM,y_svm))
```
